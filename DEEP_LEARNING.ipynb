```python
x=df.iloc[:,3:13]
y=df.iloc[:,13]
```

```python
x
```

```python
y
```

```python
geography=pd.get_dummies(x['Geography'],dtype=int)
# get values in 0 and 1
```

```python
gender=pd.get_dummies(x['Gender'],dtype=int)
```

```python
#concatenate these above variables with these datasets
x.drop(['Geography','Gender'],axis=1,inplace=True)
```

```python
x=pd.concat([x,geography,gender],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
##part 2 now lets create ANN
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ReLU
```

```python
from tensorflow.keras.layers import Dropout
```

```python
 #initialise the ANN
```

```python
clf=Sequential()
```

```python
clf.add(Dense(units=11,activation='relu'))
```

```python
#adding the first hidden layer

clf.add(Dense(units=7,activation='relu'))
```

```python
#adding the 2nd hidden layer
clf.add(Dense(units=6,activation='relu'))
```

```python
##Adding the output layer

clf.add(Dense(1,activation='sigmoid'))
```

```python
clf.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#by default the adam optimiser uses the learning rate of 0.01
```

```python
import tensorflow
tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# early stopping

import tensorflow as tf
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
)
```

```python
model=clf.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=1000)
```
