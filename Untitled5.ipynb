```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalised"]
```

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

```python
#NLTK has RegexpStemmer class which basically takes a singular regular expression and removes any prefix or suffix.
from nltk.stem import RegexpStemmer
reg=RegexpStemmer('ing$|s$|e$|able$|',min=4)
reg.stem('eating')

```

```python
reg.stem('ingeating')
```

```python
reg=RegexpStemmer('$ing|s$|e$|able$|',min=4)
reg.stem('ingeating')
```
